# Mathematical Functionality

```python
def high(Overall,df):
  if "Overall" in df.columns:
    sorted_values = df["Overall"].sort_values()
    max_value = sorted_values.iloc[0]
    second_max_value = sorted_values.iloc[1]
    difference = max_value - second_max_value
    return (f"The difference between the max and second max values is: {difference}")
```

```python
print(high("Overall",df))
```

```python
def std(df):
  std_dev = df['Overall'].std()
  return (std_dev)
```

```python
print(std(df))
```

```python
def sum(df):
  greater = df[df['Joined'] > 100]['Overall'].sum()
  return (greater)
```

# Database : SQLite3

```python
import sqlite3
import json
```

```python
json_file = "/content/eda_insights.json"
with open(json_file, 'r') as f:
    eda_insights = json.load(f)
shape_df = pd.DataFrame([eda_insights['shape']], columns=['rows', 'columns'])
columns_df = pd.DataFrame(eda_insights['columns'], columns=['Joined'])
datatypes_df = pd.DataFrame(list(eda_insights['datatypes'].items()), columns=['Joined', 'datatype'])
missing_values_df = pd.DataFrame(list(eda_insights['missing_values'].items()), columns=['Joined', 'missing_count'])
summary_df = pd.DataFrame(eda_insights['summary'])
```

```python
conn = sqlite3.connect("data.db")

shape_df.to_sql("shape_table", conn, if_exists="replace", index=False)
columns_df.to_sql("columns_table", conn, if_exists="replace", index=False)
datatypes_df.to_sql("datatypes_table", conn, if_exists="replace", index=False)
missing_values_df.to_sql("missing_values_table", conn, if_exists="replace", index=False)
summary_df.to_sql("summary_table", conn, if_exists="replace", index=False)

```

```python
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
for table in tables:
  print(table[0])
```
