```python
profile = yd(df, title="EDA Report", explorative=True)
profile.to_file("eda_report.html")
insights = {
        "shape": df.shape,
        "columns": list(df.columns),
        "datatypes": df.dtypes.astype(str).to_dict(),
        "missing_values": df.isnull().sum().to_dict(),
        "summary": df.describe(include="all").to_dict(),
    }
with open("eda_insights.json", "w") as f:
  json.dump(insights, f, indent=4)
```

```python
def chatbot_with_ydata(query, eda_insights):
    if "shape" in query:
        return f"The dataset has {eda_insights['shape'][0]} rows and {eda_insights['shape'][1]} columns."
    elif "missing" in query:
        return f"Missing values: {eda_insights['missing_values']}"
    elif "summary" in query:
        return f"Summary Statistics: {eda_insights['summary_stats']}"
    elif "columns" in query:
        return f"The dataset has the following columns: {', '.join(eda_insights['columns'])}."
    else:
        return "I'm not sure how to answer that. Can you rephrase?"
```

```python
with open("eda_insights.json", "r") as f:
    eda_insights = json.load(f)
```

```python
while True:
  query = input("You: ")
  if "exit" in query.lower():
    print("Goodbye!")
    break
  else:
    print(chatbot_with_ydata(query, eda_insights))
```

# Database : ChromaDB

```python
import chromadb
```

```python
def load_to_chromdb():
  client = chromadb.Client()
  collection_name = "mycol"
  collection = client.create_collection(name=collection_name)
  collection.add(
            ids=["eda_insights"],
            metadatas={"type": "eda_report"},
            documents=[json.dumps(insights)],
        )
  results = collection.query(query_texts=["embeddings"], n_results=5)
  return results
```

```python
def chatbot(query):
    try:
        result = collection.query(
            query_texts=[query],
            n_results=1,
        )
        if result["documents"]:
            eda_insights = json.loads(result["documents"][0])
            if "shape" in query:
                return f"The dataset has {eda_insights['shape'][0]} rows and {eda_insights['shape'][1]} columns."
            elif "missing" in query:
                return f"Missing values: {eda_insights['missing_values']}"
            elif "columns" in query:
                return f"The dataset contains the following columns: {', '.join(eda_insights['columns'])}."
            else:
                return "rephrase?"
        else:
            return "No relevant data."
    except Exception as e:
        return f"Error querying Chroma: {e}"
```

```python
json_string = json.dumps(eda_insights, indent=4)
print(json_string)
```

```python
while True:
    query = input("You: ")
    if "exit" in query.lower():
        print("Goodbye!")
        break
    else:
        print(chatbot(query))
```
